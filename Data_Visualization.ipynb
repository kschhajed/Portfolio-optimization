```python
from pandas_datareader.yahoo.daily import YahooDailyReader
import pandas as pd
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stocks = ['NOVA', 'BEPC', 'CIO', 'JOE', 'CC', 'KWR', 'NET', 'BTRS', 'DXCM', 'MNKD', 'AJRD', 'CARR', 'AUB', 'UWMC',
          'CPE', 'FRO', 'TTCF', 'TSN', 'F', 'LNW', 'LUMN', 'GOGO']

etfs_germany = ['ETLX.DE', 'IBCI.DE', 'G2X.DE', '2B78.DE', 'XDW0.DE', 'VOOM.DE', 'IS0D.DE', 'FLXG.DE',
                'SPYY.DE', 'IUSL.DE', 'XDEB.DE', 'EUNL.DE', 'XDWD.DE', 'XDEQ.DE', 'EMWE.DE', 'USPY.DE',
                'WTEM.DE', 'XDWH.DE', 'XDWT.DE ']

etfs_usa = ['XLY', 'XLP', 'XLE', 'XLF', 'XLV', 'XLI', 'IYE', 'ITB', 'IYR', 'XLB', 'KBE', 'XHB', 'XME', 'XOP', 'KRE',
            'XTL', 'XTL', 'XLK', ' XLU', 'GDX', 'GDXJ', 'OIH', 'SMH', 'VNQ']

start_date = '01/01/2020'
end_date = '01/03/2022'
```

```python
names = []
sector = []

for ticker in stocks:
    data = yf.Ticker(ticker)
    names.append(data.info['longName'])
    sector.append(data.info['sector'])

df_stocks = pd.DataFrame(list(zip(stocks, names, sector)), columns=['Ticker', 'Full Name', 'Sector'])
df_stocks
```

```python
names = []
sector = []

for ticker in etfs_germany:
    data = yf.Ticker(ticker)
    names.append(data.info['longName'])
#     sector.append(data.info['sector'])

df_etfs_de = pd.DataFrame(list(zip(etfs_germany, names)), columns=['Ticker', 'Full Name'])
df_etfs_de
```

```python
names = []
sector = []

for ticker in etfs_usa:
    data = yf.Ticker(ticker)
    names.append(data.info['longName'])
#     sector.append(data.info['sector'])

df_etfs_usa = pd.DataFrame(list(zip(etfs_usa, names)), columns=['Ticker', 'Full Name'])
df_etfs_usa
```

```python
data = YahooDailyReader(
    symbols=stocks,
    start='01/01/2020', end='01/03/2022',
    retry_count=3, pause=0.1, session=None, adjust_price=False, ret_index=False,
    chunksize=1, interval='d', get_actions=False, adjust_dividends=True)

all_data_s = data.read()
adj_close_prices_s = all_data_s['Adj Close']

adj_close_prices_s.tail()
```

```python
adj_close_prices_s.index
```

```python
plot = adj_close_prices_s.plot(figsize=(15,10))
fig = plot.get_figure()
fig.savefig("results/all_stocks_prices.png")
```
